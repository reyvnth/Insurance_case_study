```
def percentile_based_outlier(data, threshold=95):
    diff = (100 - threshold) / 2.0
    minval, maxval = np.percentile(data, [diff, 100 - diff])
    return (data < minval) | (data > maxval)

col_names = ['Number of Cars','Average Age of Car(s)', 'Number of Trips']

fig, ax = plt.subplots(len(col_names), figsize=(15,8))

for i, col_val in enumerate(col_names):
    x = data[col_val][:1000]
    sn.distplot(x, ax=ax[i], rug=True, hist=False)
    outliers = x[percentile_based_outlier(x)]
    ax[i].plot(outliers, np.zeros_like(outliers), 'ro', clip_on=False)

    ax[i].set_title('Outlier detection - '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    
plt.show()
```

```
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```
print((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR)))
```
